## DATASET 2 FROM DATACAMP(Hkust Dataset)

```python
data3=pd.read_csv("en_dataset.csv")
data3
```

```python
mask=data3.sentiment=='offensive'
mask1=data3.sentiment=='hateful'
mask2=data3.sentiment=='normal'
```

```python
new_data3=data3[mask | mask1 | mask2]
```

```python
new_data3.sentiment.value_counts()
```

```python
new_data3.rename(columns={'sentiment':'label'},inplace=True)
```

```python
def change_label(text):
    if text=='offensive':
        return 1
    elif text=='hateful':
        return 0
    else:
        return 2
```

```python
new_data3.label=new_data3.label.map(change_label)
```

```python
new_data3.label.value_counts()
```

```python
new_data3=new_data3[['tweet','label']]
new_data3
```

# SMOKING DATASET

```python
data7=pd.read_csv('gold_corpus.csv')
data7
```

```python
data7.tweet=data7.tweet.str.replace("\$RESERVED\$\$MENTION\$", '@$RESERVED$ $MENTION$',regex=True)
data7.tweet=data7.tweet.str.replace("\$MENTION\$", '@$MENTION$',regex=True)
data7.tweet=data7.tweet.str.replace("\$NUMBER\$ ", '#$NUMBER$',regex=True)
data7
```

```python
data7.tweet=data7.tweet.str.replace("\$URL\$ ", '$URL.com$',regex=True)
```

```python
data7.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4',
       'Unnamed: 5', 'Unnamed: 6'],axis=1,inplace=True)
```

```python
data7.label=3
```

```python
data7.label.value_counts()
```

## DATASET 6 From Kaggle

```python
data_rt=pd.read_fwf("train_text.txt", sep=" ",header=None)
data_rt.rename(columns={0:"tweet"},inplace=True)
data_rt
```

```python
data_rt_test=pd.read_fwf("test_text.txt", sep=" ",header=None)
data_rt_test.rename(columns={0:"tweet"},inplace=True)
data_rt_test
```

```python
data_rt_label=pd.read_fwf("train_labels.txt", sep=" ",header=None)
data_rt_label.rename(columns={0:"label"},inplace=True)
data_rt_label
```

```python
data_rt_test_label=pd.read_fwf("test_labels.txt", sep=" ",header=None)
data_rt_test_label.rename(columns={0:"label"},inplace=True)
data_rt_test_label
```

```python
data_hate=pd.concat([data_rt,data_rt_label],axis=1)
data_hate
```

```python
data_hate.dropna(inplace=True)
data_hate
```

```python
data_hate_test=pd.concat([data_rt_test,data_rt_test_label],axis=1)
data_hate_test
```

```python
data_hate_test.dropna(inplace=True)
```

```python
rt_data=pd.concat([data_hate,data_hate_test])
rt_data
```

```python
mask=rt_data.label==1
kr=rt_data[mask]
```

```python
kr.label=0
kr
```
